```python
# pip install matplotlib pandas

import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import matplotlib.dates as mdates

# ---- Work Experience with Category ----
roles = [
    ("Tabloid Editor  The Diamondback", "2023-04", "2024-06", "Campus Job"),
    ("Student Office Assistant  Merrill College of Journalism", "2023-08", "Present", "Campus Job"),
    ("Member  Phi Chi Theta (PCT)", "2023-11", "2024-05", "Leadership"),
    ("Blockchain & Web3 Security Extern  Webacy", "2024-03", "2024-04", "Internship"),
    ("DEI Committee Head  PCT", "2024-05", "2024-12", "Leadership"),
    ("Vice President  The Diabetes Link", "2024-05", "Present", "Leadership"),
    ("Data Intern  Colonial Parking", "2024-06", "2024-08", "Internship"),
    ("Cognizant Leadership Summit 2024", "2024-08", "2024-08", "Event"),
    ("Director of DEI  PCT", "2024-12", "Present", "Leadership"),
]

# ---- Color Palette ----
category_colors = {
    "Internship": "#4C78A8",  # blue
    "Leadership": "#72B7B2",  # green
    "Event": "#E15759",       # red
    "Campus Job": "#F28E2B",  # orange
}

# ---- Convert Dates ----
def to_dt(date_str):
    if date_str == "Present":
        return datetime.today()
    return datetime.strptime(date_str, "%Y-%m")

df = pd.DataFrame(roles, columns=["Role", "Start", "End", "Category"])
df["Start"] = df["Start"].apply(to_dt)
df["End"] = df["End"].apply(to_dt)

# ---- Sort by Start Date ----
df = df.sort_values("Start", ascending=True)
y_pos = range(len(df))

# ---- Plot Horizontal Bars ----
fig, ax = plt.subplots(figsize=(12, 6))
for i, row in df.iterrows():
    ax.barh(
        y=i,
        width=(row["End"] - row["Start"]).days,
        left=mdates.date2num(row["Start"]),
        color=category_colors[row["Category"]],
        edgecolor="black",
        label=row["Category"] if row["Category"] not in ax.get_legend_handles_labels()[1] else ""
    )

# ---- Y-Axis Labels ----
ax.set_yticks(y_pos)
ax.set_yticklabels(df["Role"], fontsize=9)

# ---- X-Axis Formatting ----
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
fig.autofmt_xdate()

# ---- Title & Legend ----
ax.set_xlabel("Timeline", fontsize=10)
ax.set_title("Sudenur Yavuz  Work Experience Overview", fontsize=13, weight="bold")
ax.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(axis="x", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
